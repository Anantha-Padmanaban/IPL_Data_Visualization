## IPL Data Analysis and Visualization

## Importing Packages and DataSets 

```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import seaborn as sns
sns.set_style('darkgrid')

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter("ignore")
```

```python
''' reading and merging datasets from CSV file'''
match_data = pd.read_csv("C:/Users/VSmart/IPL_Data_Visualization/IPLMatches2008-2020.csv")

ball_data = pd.read_csv("C:/Users/VSmart/IPL_Data_Visualization/IPL Ball-by-Ball 2008-2020.csv")

ipl_DataSets = pd.merge(ball_data,match_data, left_on='id', right_on ='id')
```

## Visualization

## Winners of IPL across all seasons

```python
#creating a dataframe with season and winner columns
winning_teams = ipl_DataSets[['season','winner']]

#dictionaries to get winners to each season
winners_team = {}
for i in sorted(winning_teams.season.unique()):
    winners_team[i] = winning_teams[winning_teams.season == i]['winner'].tail(1).values[0]
    
winners_of_IPL = pd.Series(winners_team)
winners_of_IPL = pd.DataFrame(winners_of_IPL, columns=['team'])
winners_of_IPL['team'].value_counts().plot(kind = 'barh', figsize = (15,5), color = 'red')
plt.title("Winners of IPL across all seasons",fontsize=18,fontweight="bold")
plt.ylabel("Teams", size = 25)
plt.xlabel("count", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

## Total Number of Matches Won by each team

```python
winners = match_data.winner.value_counts()
plt.figure(figsize=(20, 7))
sns.barplot(winners.index, winners)
plt.xlabel('Team')
plt.xticks(rotation=90)
plt.ylabel('Count');
plt.title('Total Number of Matches Won by each team', fontdict={'fontsize': 20, 'fontweight': 'medium'});
```

## Number of matches played each season

```python
plt.figure(figsize=(15,5))
sns.countplot('season', data = match_data)
plt.title("Number of matches played each season",fontsize=18,fontweight="bold")
plt.ylabel("Count", size = 25)
plt.xlabel("Season", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

## Number of teams participated each season

```python
ipl_DataSets.groupby('season')['team1'].nunique().plot(kind = 'bar', figsize=(15,5), color = 'black')
plt.title("Number of teams participated each season ",fontsize=18,fontweight="bold")
plt.ylabel("Count of teams", size = 25)
plt.xlabel("Season", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

## Number of Matches Played in each Cities

```python
''' number of matches played in each cities '''
city = match_data['city'].value_counts().sort_values().nlargest(10)
plt.figure(figsize=(15, 5))
sns.barplot(city.index, city)
plt.ylabel('City')
plt.xlabel('Count of Number Matches Played');
plt.xticks(rotation=45)
plt.title("Number of Matches Played in each Cities",fontsize=18,fontweight="bold")
```

## Number of Matches Played in each Venue

```python
''' number of matches played in each Venue'''
city = match_data['venue'].value_counts().sort_values().nlargest(10)
plt.figure(figsize=(15, 5))
sns.barplot(city.index, city)
plt.ylabel('City')
plt.xlabel('Count of Number Matches Played');
plt.xticks(rotation=90)
plt.title("Number of Matches Played in each Venue",fontsize=18,fontweight="bold")
```

## Total Number of Toss Won by each Team

## Total runs scored by each Team

```python
temp_df = ipl_DataSets.groupby('batting_team')['batsman_runs'].agg('sum').reset_index().reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]
plt.subplots(figsize=(13,8))
sns.barplot(x='batting_team',y='batsman_runs',data=temp_df)

plt.title('Total runs scored by each Team',fontsize=13)
plt.xticks(rotation=90)
```

## Average Runs scored by a batting team each season

```python
innings_score = ipl_DataSets.groupby(['id', 'season', 'batting_team'])['total_runs'].sum().reset_index()
season_avg_score = innings_score.groupby(['season'])['total_runs'].mean().reset_index()
#bar plot
plot = sns.barplot(x='season', y='total_runs', data=season_avg_score)
plot.set(xlabel='Season', ylabel='Total runs')
plot.set_title('Average score by a batting team each season')
plt.show()
```

## No's of 4 and 6 scored by each Team

```python
fours = ipl_DataSets.groupby(['batting_team'])['batsman_runs'].agg(lambda x:(x==4).sum()).reset_index().rename(columns={'batsman_runs':'4s'})
sixes = ipl_DataSets.groupby(['batting_team'])['batsman_runs'].agg(lambda x:(x==6).sum()).reset_index().rename(columns={'batsman_runs':'6s'})
boundaries = fours.merge(sixes,left_on='batting_team',right_on='batting_team',how='left')
#print(fours)
#print(sixes)
#print(boundaries)
plot = boundaries.set_index('batting_team').plot(kind='bar',figsize=(13,8))
plt.xticks(np.arange(0,13,1))
plot.set_xticklabels=['CSK','DC','DD','GL','KXIP','KTK','MI','PW','RPS','RCB','SRH']
plt.yticks(np.arange(0,2201,200))
plt.xlabel('')
plt.show()
```

## Top 10 Batsman 

```python
#batsman with most runs
temp_df = ipl_DataSets.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]
plt.subplots(figsize=(13,8))
sns.barplot(x='batsman',y='batsman_runs',data=temp_df)
plt.yticks(np.arange(0,5000,500))
plt.title('BATSMAN WITH MOST RUNS',fontsize=13)
```

## Average run scored by each Team in a Match

```python
df = ipl_DataSets.groupby(['id','batting_team'])['total_runs'].sum().reset_index()
df = df[df['total_runs'] > 40]
df = df.sort_values(by=['batting_team'])
plt.subplots(figsize=(15,10))
plt.xticks(rotation=90)
plot = sns.barplot(x='batting_team',y = 'total_runs',data=df,palette=sns.color_palette('Paired',14))
# plot.set_title(team,fontsize=15)
plot.set_xlabel('TEAM',fontsize = 10)
plot.set_ylabel('RUNS SCORED',fontsize=10)
plot.set_title('Average run scored by each Team in a Match',fontsize=15)
```

## Top 10 Bowlers

## Comparing Two Teams

```python
#compare two teams
def team_comparator(team1,team2):
    compare = ipl_DataSets[((ipl_DataSets['team1']==team1) | (ipl_DataSets['team2']==team1)) & ((ipl_DataSets['team1']==team2) | (ipl_DataSets['team2']==team2))]
    plt.subplots(figsize=(13,8))
    plt.yticks(np.arange(0,4,1))
    plot = sns.countplot(x='season', hue='winner',data = compare)
    plot.set_xlabel('SEASON',fontsize=10)
    plot.set_ylabel('WIN COUNT',fontsize=10)
team_comparator('Chennai Super Kings','Royal Challengers Bangalore')
```
